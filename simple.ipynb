```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the loss function (e.g., f(x, y) = x^2 + y^2)
def loss_function(x, y):
    return x**2 + y**2

# Define the gradient of the loss function
def gradient(x, y):
    return 2*x, 2*y

# Gradient descent parameters
learning_rate = 0.1
iterations = 20

# Starting point
x, y = 2.0, 2.0  # Initial values
path = [(x, y)]  # To store the path of descent

# Perform gradient descent
for _ in range(iterations):
    grad_x, grad_y = gradient(x, y)
    x -= learning_rate * grad_x
    y -= learning_rate * grad_y
    path.append((x, y))

# Convert path to numpy arrays for plotting
path = np.array(path)
x_path, y_path = path[:, 0], path[:, 1]
z_path = loss_function(x_path, y_path)

# Create a grid for the surface plot
x_vals = np.linspace(-2.5, 2.5, 100)
y_vals = np.linspace(-2.5, 2.5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = loss_function(X, Y)

# Plot the 3D surface and gradient descent path
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Gradient descent path
ax.plot(x_path, y_path, z_path, color='r', marker='o', label='Gradient Descent Path')

# Labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Loss')
ax.set_title('3D Gradient Descent Visualization')
ax.legend()

plt.show()
```
